```python
import random
```

```python
responses = {

    "hi": [
        "Hello!",
        "Hey there!",
        "Hi student 👋"
    ],

    "ml": [
        "Machine Learning is AI that learns from data.",
        "ML helps systems recognize patterns."
    ],

    "robotics": [
        "Robotics combines AI with machines.",
        "Robots use sensors and intelligent control."
    ],

    "bye": [
        "Goodbye!",
        "See you later!",
        "Keep building cool AI stuff 🚀"
    ]
}
```

```python
user_input = input("You: ").lower()
```

```python
if user_input in responses:

    print(
        "Bot:",
        random.choice(responses[user_input])
    )

else:

    print(
        "Bot: Sorry, I don't understand."
    )
```

```python
while True:

    user_input = input("You: ").lower()

    if user_input == "bye":

        print("Bot:", random.choice(responses["bye"]))

        break

    elif user_input in responses:

        print(
            "Bot:",
            random.choice(responses[user_input])
        )

    else:

        print(
            "Bot: Sorry, I don't understand."
        )
```
